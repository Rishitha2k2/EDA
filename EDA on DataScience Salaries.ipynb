# HANDLING MISSING VALUES

##### USE OF NUMPY:NumPy can be used to perform a wide variety of mathematical operations on arrays. It adds powerful data structures to Python that guarantee efficient calculations with arrays and matrices and it supplies an enormous library of high-level mathematical functions that operate on these arrays and matrices.

```python
df2=df.replace(to_replace=np.nan,value=0)
df2
```

```python
df3=df.fillna(value=5)
df3
```

```python
#Using pas method to fill null values
df4=df.fillna(method='pad')
df4
```

```python
#Using bfill method to fill null values
df5=df.fillna(method='bfill',axis=0)
df5
```

```python
#Filling the remote_ratio column with mean value of that column
df6=df.fillna(value=df['remote_ratio'].mean())
df6
```

```python
#Filling the work_year column with maximum value
df7=df.fillna(value=df['work_year'].max())
df7
```

# DATA VISUALISATION

##### USE OF MATPLOTLIB:Matplotlib is a python library used to create 2D graphs and plots by using python scripts. It has a module named pyplot which makes things easy for plotting by providing feature to control line styles, font properties, formatting axes etc

```python
#Line graph
plt.plot(df.experience_level,df.salary);
plt.title("EXPERIENCE LEVEL VS SALARY")
```

##### USE OF SEABORN:Seaborn helps you explore and understand your data. Its plotting functions operate on dataframes and arrays containing whole datasets and internally perform the necessary semantic mapping and statistical aggregation to produce informative plots.

```python
#Scatter plot
sns.scatterplot(df.experience_level,df.salary,s=50,hue=df.experience_level);
plt.figure(figsize=(12,13))
```

```python
#BARCHART
barplot=df.groupby('work_year')[['remote_ratio']].mean();
plt.bar(barplot.index,barplot.remote_ratio);
plt.title('WORK YEAR vs REMOTE RATIO')
```

```python
#PIE CHART FOR EXPERIENCE LEVEL
plt.title("PIE CAHRT FOR EXPERIENCE LEVEL ")
df['experience_level'].value_counts().plot(kind="pie", autopct="%.2f",radius=1.2)
plt.show();
```

```python
x=df.work_year
y=df.salary_in_usd
plt.scatter(x,y)
plt.xlabel('work_year')
plt.ylabel('salary_in_usd')
#plt.title('')
plt.show();
```
